```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Importing movies.csv to python

df = pd.read_csv(r'C:\Users\Ar_Da\OneDrive\Desktop\SQL portfolio project\PythonProjects\movies.csv')

df.head()
```

```python

## DATA CLEANING
# Checking if there is a missing data

df.isnull().sum()
```

```python
# drop null

df=df.dropna()

df.isnull().sum()
```

```python
#Checking data types for our columns

df.dtypes
```

```python
#Changing datatypes

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

#drop any duplicates

df=df.drop_duplicates()
```

```python
##DATA ANALYSIS
#See if there is a correlation between budget and gross

plt.scatter(x=df['budget'], y=df['gross'], s=10)
plt.title('Gross Earning VS Budget')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earning')
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red', 's':10})
plt.show()
```

```python
#Find the pearson r value to put in the graph

r = df['budget'].corr(df['gross'], method='pearson')

r
```

```python
#Look at the correlation matrix to see correlation between other factors

correlation_matrix  = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.show()
```

```python
#We see there is a strong positive correlation between gross and budget
#Now we put the pearson r value in the graph to show correlation

plt.scatter(x=df['budget'], y=df['gross'], s=10)
plt.title('Gross Earning VS Budget')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earning')
plt.annotate('pearson r = 0.74', xy=(0.5e+8,2.5e+9), fontsize = 15)
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red', 's':10})
plt.show()
```
